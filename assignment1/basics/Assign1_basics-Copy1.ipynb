```python
def finalist(datlist):
    finalist=[]
    for i in range(len(datlist)):
        str=datlist[i]
        count=0
        for j in range(len(str)):
            for k in range(j+1,len(str)):
                if str[j]==str[k]:
                    count=1
        if count==0: finalist.append(str)
    return(finalist)
```

```python
def ldim(word,aset):
    count=0
    branches=[]
    for test in aset:
        branch=cd(word,test)
        if branch not in branches:
            count+=1
            branches.append(branch)
    return count        
```

```python
# import functools
# @functools.lru_cache(maxsize=1280000)
def best_choice(aset):
    max_branch=0
    word=aset[0]# default just in 
    for test in aset:
        count=ldim(test,aset)
        if count>max_branch:
            word=test
            max_branch=count
    return word
```

```python
import pickle

from functools import wraps
def memoize(func):
    cache = {}
    @wraps(func)
    def wrap(*args,**kwargs):
        key = pickle.dumps((args, kwargs))
        if key not in cache:
            print ('Running func with ', args)
            cache[key] = func(*args, **kwargs)
        else:
            print ('result in cache')
        return cache[key]
    return wrap
```
